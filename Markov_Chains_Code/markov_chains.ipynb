## Computing the Transition Matrix

```python
counter = 0
word_index = {}
for i in list(set(unique_chars)):
  word_index[i] = counter
  counter += 1
total_word = len(set(unique_chars))
transition_matrix = np.zeros((total_word,total_word))
for i in range(len(unique_chars) -1 ):
  transition_matrix[ word_index[unique_chars[i]] ][ word_index[unique_chars[i+1]] ] += 1
```

```python
row_sums = transition_matrix.sum(axis=1)
transition_matrix = transition_matrix / row_sums[:, np.newaxis]
```

```python
import numpy as np

def generate_markov_chain(transition_matrix, initial_state, num_steps):
    num_states = len(transition_matrix)
    current_state = initial_state
    markov_chain = [current_state]

    for _ in range(num_steps):
        probabilities = transition_matrix[current_state]
        next_state = np.random.choice(num_states, p=probabilities)
        markov_chain.append(next_state)
        current_state = next_state

    return markov_chain

# Example usage
# transition_matrix = np.array([[0.7, 0.3], [0.4, 0.6]])  # Transition matrix

initial_state = word_index["b"]  # Initial state
num_steps = 150  # Number of steps to generate
markov_chain = generate_markov_chain(transition_matrix, initial_state, num_steps)
print(markov_chain)
```
